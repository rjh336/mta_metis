```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cleaned_df = pd.read_csv('./prepped_data_2015-2017_presummer.csv')
```

```python
cleaned_df.head()
```

```python
cleaned_df.info()
```

```python
# Datetimes get converted to strings upon import, so we convert them back

cleaned_df['DATETIME'] = pd.to_datetime(cleaned_df.DATETIME, 
                                    format="%Y/%m/%d %H:%M:%S")
cleaned_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
def top_10_stations(df):
    """ Return dataframe with top ten stations ranked by average daily TOTAL_TRAFFIC
    """
    df1 = df.groupby(['STATION', pd.Grouper(key='DATETIME', freq='1D')])\
    .sum()\
    .reset_index()\
    .groupby(['STATION'])\
    .mean()\
    .TOTAL_TRAFFIC\
    .sort_values(ascending=False)
    stations = df1.reset_index().STATION.unique()[:10]
    df2 = df[df.STATION.apply(lambda x: x in stations)][['STATION', 'DATETIME', 'TOTAL_TRAFFIC', 'DOW','TIME']]
    return df2

def total_traffic_weekday_box(df, title):
    """ Return boxplots showing distribution of average daily TOTAL_TRAFFIC
        by day of the week
    """
    df_pre = df[['DATETIME', 'DOW', 'TOTAL_TRAFFIC', 'STATION']]
    df1 = df_pre.groupby([pd.Grouper(key='DATETIME', freq='1D'), 'DOW', 'STATION'])\
    .sum()\
    .reset_index()
    #print(df1.head(10))

    fig, ax1 = plt.subplots(figsize=(15,8))
    ax = sns.boxplot(x='DOW', y='TOTAL_TRAFFIC', data=df1, showfliers=False, ax=ax1)
    ax.set_ylabel('Daily Traffic Across Stations')
    ax.set_xlabel('Day of the Week')
    ax.set_title(title)
    fig.show()
    
def average_hourly_overday(df, n_stations = 10):
    """ Return line chart that shows average daily TOTAL_TRAFFIC
        by TIME (i.e. 4-hour time intervals)
    """
    # Take only relavent columns
    df_pre = df[['STATION', 'DATETIME', 'TOTAL_TRAFFIC', 'TIME']] 
    
    # Group into dailies through sum and then find the average for stations per day
    df1 = df_pre.groupby(['STATION', pd.Grouper(key='DATETIME', freq='1D')])\
    .sum()\
    .reset_index()\
    .groupby(['STATION'])\
    .mean()\
    .TOTAL_TRAFFIC\
    .sort_values(ascending=False)[0:n_stations]
    
    # Collection station labels
    stations = df1.reset_index().STATION.unique()
    
     # Create new dataframe 
    df2 = df_pre[df.STATION.apply(lambda x: x in stations)].copy()
    
    # Create time columnn
    df3 = df2.groupby(['STATION', 'DATETIME', 'TIME'])\
    .sum()\
    .reset_index()\
    .groupby(['STATION', 'TIME'])\
    .mean()\
    .reset_index()
    #print(df3)
    # Plot
    fig, ax = plt.subplots(figsize=(15,8))
    labels = []
    for key, grp in df3.groupby('STATION'):
        labels.append(key)
        _ = grp.plot(ax=ax, x='TIME', y='TOTAL_TRAFFIC')

    lines, _ = ax.get_legend_handles_labels()
    ax.legend(lines, labels, loc='best')
    #ax.set_xticks(df3.TIME.tolist())
    ax.set_ylabel('Volume (entry + exit)')
    ax.set_xlabel('Time')
    ax.set_title('Average Volume for Top 10 Busiest Stations')
    fig.show()
    
def ten_station_hist(df):
    """ Returns top ten stations by average daily volume as a bar chart
    """
    
    df_pre = df[['STATION', 'DATETIME', 'TOTAL_TRAFFIC']]
    df1 = df_pre.groupby(['STATION', pd.Grouper(key='DATETIME', freq='1D')])\
        .sum()\
        .reset_index()\
        .groupby(['STATION'])\
        .mean()\
        .TOTAL_TRAFFIC\
        .sort_values(ascending=False)[:10]
    print(df1.shape)
    
    fig = plt.figure(figsize=(12,5))
    ax1 = fig.add_axes([0,0,1,1])
    df1.plot(kind='bar', ax=ax1)
    ax1.set_xlabel('Station')
    ax1.set_ylabel('Daily Average Volume')
    ax1.set_title('Daily Weekday Average Volume of the 10 Busiest Stations')
    return df1

def recommendation_chart(df):
    """ Returns bar chart of top ten (STATION, TIME, DOW) combos
    """
    df_pre = df[['STATION', 'DATETIME', 'TOTAL_TRAFFIC', 'DOW', 'TIME']].copy()
    
    df1 = df_pre.groupby(['STATION','DATETIME','TIME','DOW'])\
    .sum()\
    .reset_index()\
    .groupby(['STATION', 'TIME', 'DOW'])\
    .mean()\
    .sort_values(by='TOTAL_TRAFFIC', ascending=False)
    
    fig, ax = plt.subplots(figsize=(12,8))
    df1[:10]['TOTAL_TRAFFIC'].sort_values(ascending=True).plot(kind='barh', ax=ax)
    ax.set_xlabel('Average Volume')
    ax.set_ylabel('Time and Place')
    ax.set_title('Busiest Subways')
    
```

```python
# TOP 10 Stations by average daily volume
top_ten = top_10_stations(cleaned_df)
top_ten.shape
```

```python
top_ten.head(10)
```

```python
ten_station_hist(top_ten)
```

```python
total_traffic_weekday_box(cleaned_df, 'All Stations')
```

```python
total_traffic_weekday_box(top_ten, "Top 10 Stations by Average Daily Volume")
```

```python
average_hourly_overday(cleaned_df, n_stations = 10)
```

```python
recommendation_chart(top_ten)
```
