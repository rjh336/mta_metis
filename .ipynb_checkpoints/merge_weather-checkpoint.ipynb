```python
# Import weather data for NYC

# All of 2016 till September 2017

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Import weather data for NYC

# All of 2016 till September 2017

df_weather = pd.read_csv("NY_weather_summer.csv")

# convert DATE to datetime

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'], format='%m/%d/%y')

# Create Flag for bad weather
# Set BAD_WEATHER equal to threshold for inches of precipitation (based on mean PRCP)
precip_threshold = .12
df_weather['BAD_WEATHER'] = np.where((df_weather['PRCP'] >= precip_threshold) | (df_weather['SNWD'] > 0), 1, 0)

drop_cols = [col for col in df_weather.columns if col[:2] == 'WT']
df_weather = df_weather.drop(drop_cols, axis=1)
df_weather.info()
#print(drop_cols)
```

```python
def ten_station_hist_rain(df):
    """ Return top 10 bar chart by STATION comparing
        when BAD_WEATHER occured vs no BAD_WEATHER
    """
    df_pre = df[['STATION', 'DATETIME', 'TOTAL_TRAFFIC', 'BAD_WEATHER']]
    df_pre_rain = df_pre[df_pre['BAD_WEATHER'] == 'Yes']
    df_pre_no_rain = df_pre[df_pre['BAD_WEATHER'] == 'No']
    
    df1_rain = df_pre_rain.groupby(['STATION', pd.Grouper(key='DATETIME', freq='1D')])\
        .sum()\
        .reset_index()\
        .groupby(['STATION'])\
        .mean()\
        .TOTAL_TRAFFIC\
        .sort_values(ascending=False)[:10]
    print(df1_rain)
    
    df1_no_rain = df_pre_no_rain.groupby(['STATION', pd.Grouper(key='DATETIME', freq='1D')])\
        .sum()\
        .reset_index()\
        .groupby(['STATION'])\
        .mean()\
        .TOTAL_TRAFFIC\
        .sort_values(ascending=False)[:10]
    print(df1_no_rain)
    
    df1 = pd.DataFrame(df1_rain)
    df1['no_rain'] = df1_no_rain
    df1 = df1.rename(index=str, columns={"TOTAL_TRAFFIC": "Rain", "no_rain": "No Rain"})
    
    fig = plt.figure(figsize=(12,5))
    ax1 = fig.add_axes([0,0,1,1])
    df1.plot(kind='bar', ax=ax1)
    #df1_no_rain.plot(kind='bar', ax=ax1)
    ax1.set_xlabel('Station')
    ax1.set_ylabel('Daily Average Volume')
    ax1.set_title('Daily Weekday Average Volume of the 10 Busiest Stations')
    fig.show()

def top_10_stations(df):
    df1 = df.groupby(['STATION', pd.Grouper(key='DATETIME', freq='1D')])\
    .sum()\
    .reset_index()\
    .groupby(['STATION'])\
    .mean()\
    .TOTAL_TRAFFIC\
    .sort_values(ascending=False)
    print(df1)
    stations = df1.reset_index().STATION.unique()[:10]
    #print(stations)
    df2 = df[df.STATION.apply(lambda x: x in stations)][['STATION', 'DATETIME', 'TOTAL_TRAFFIC', 'DOW', 'BAD_WEATHER']]
    return df2
```

```python
# convert DATETIME back to DT object
merged['DATETIME'] = merged['DATETIME'] = pd.to_datetime(merged.DATETIME, format="%Y/%m/%d %H:%M:%S")
```

```python
top_ten = top_10_stations(merged)
top_ten
```

```python
# Over three year period 2015 - 2017, rain tends to lower ridership for top 10
# Over one year period, rain tends to have no discernable impact for top 10
ten_station_hist_rain(top_ten)
```
